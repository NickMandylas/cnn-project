```python
import tensorflow as tf
from tensorflow import keras
import PIL
import numpy as np
```

```python
classes = {0: ('akiec', 'Actinic keratoses and intraepithelial carcinomae'),  
           1:('bcc' , ' basal cell carcinoma'), 
           2 :('bkl', 'benign keratosis-like lesions'), 
           3: ('df', 'dermatofibroma'),
           4: ('nv', ' melanocytic nevi'),
           5: ('vasc', ' pyogenic granulomas and hemorrhage'), 
           6: ('mel', 'melanoma')}
```

```python
# sequential 
```

```python
sequential= tf.keras.models.load_model('gdrive/Shareddrives/FIT3163 - Group 6 Assignment/CNN/functional_correct.h5')
```

```python
image = PIL.Image.open('gdrive/Shareddrives/FIT3163 - Group 6 Assignment/CNN/ISIC_0024307.jpg')
image = image.convert("RGB")
image = image.resize((28,28))
img = np.array(image).reshape(-1,28,28,3)
result = sequential.predict(img)
```

```python
PIL.Image.open('gdrive/Shareddrives/FIT3163 - Group 6 Assignment/CNN/ISIC_0024307.jpg')
```

```python
max_prob = max(result[0])
class_ind = list(result[0]).index(max_prob)
class_label = classes[class_ind]
print(class_label)
```
